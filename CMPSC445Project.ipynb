# Preprocessing the Dataset

```python
#find the nullvalues
df.isnull().sum()
```

```python
df['Gender'].mode()[0]
```

```python
#Replace all null values with new data
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mean())
```

```python
#fill the missing values for categorical terms with mode of the values of the attribute.
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])

```

```python
#find the nullvalues
df.isnull().sum()
```

```python
df['ApplicantIncome']= df['ApplicantIncome'].replace(0,df['ApplicantIncome'].mean())
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
cols = ['Gender', 'Married', "Education", "Self_Employed", "Property_Area", "Dependents"]
#cols = ["Dependents"]
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

# Train_Test Split

```python
#specify input and output attributes
X = df.drop(columns=['Loan_Status'], axis=1)
y = df['Loan_Status']
```

```python
data = X.values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Train the Model

```python
#classify function
from sklearn.model_selection import cross_val_score
def classify(model, x, y):
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    model.fit(x_train, y_train)
    print("Accuracy is", model.score(x_test, y_test) * 100)
    #cross validation in the entire dataset
    score = cross_val_score(model, x, y, cv=5)
    print("Cross validation is, ", np.mean(score) * 100)
```

```python
#let's run the classifier with the basic model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model, X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
classify(model, X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
model = RandomForestClassifier()
classify(model, X, y)
```

```python
model = ExtraTreesClassifier()
classify(model, X, y)
```

# Hyperparameter tuning(Changing the hyperparameter for better accuracy)

```python
model = RandomForestClassifier(n_estimators = 100, min_samples_split = 25, max_depth = 7, max_features = 1)
classify(model, X, y)
```
